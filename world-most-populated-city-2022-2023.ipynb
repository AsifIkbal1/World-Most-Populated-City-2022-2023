```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
```

```python
df_asia = df[df["Continent"] == "Asia"]
df_s_america = df[df["Continent"] == "South America"]
df_n_america = df[df["Continent"] == "North America"]
df_europe = df[df["Continent"] == "Europe"]
df_africa = df[df["Continent"] == "Africa"]
df_oceania = df[df["Continent"] == "Oceania"]
```

```python
df_grouped_22 = df[["Continent", "Pop2022"]].groupby("Continent").sum().sort_values(by="Pop2022", ascending=False)
df_grouped_23 = df[["Continent", "Pop2023"]].groupby("Continent").sum().sort_values(by="Pop2023", ascending=False)

print(df_grouped_22,"\n","\n", df_grouped_23)
```

```python
df_sorted_asia_22 = df_asia[["City", "Pop2022"]].sort_values(by="Pop2022", ascending=False).head(10).reset_index(drop=True)
plt.figure(figsize=(10, 6))
plt.bar(df_sorted_asia_22["City"], df_sorted_asia_22["Pop2022"])
```

<div style="padding:20px;

```python
df_sorted_asia_23 = df_asia[["City", "Pop2023"]].sort_values(by="Pop2023", ascending=False).head(10).reset_index(drop=True)
plt.figure(figsize=(10, 6))
plt.bar(df_sorted_asia_23["City"], df_sorted_asia_23["Pop2023"])
```

<a id="1"></a>
<div style="padding:20px;
            color:white;
            margin:10;
            font-size:150%;
            text-align:left;
            display:fill;
            border-radius:5px;
            background-color:#294B8E;
            overflow:hidden;
            font-weight:700">Analysis of population difference between 2022 and 2023 in Asia</div>

```python
df_asia["year_diff"] = df_sorted_asia_23["Pop2023"] - df_sorted_asia_22["Pop2022"]
print(df_asia["year_diff"])

df_sorted_asia = df_asia[["City", "year_diff"]].sort_values(by="year_diff", ascending=False).head(10).reset_index(drop=True)
plt.figure(figsize=(10, 6))

plt.bar(df_sorted_asia["City"], df_sorted_asia["year_diff"])
for i in range(len(df_sorted_asia)):
    plt.text(df_sorted_asia.index[i], df_sorted_asia["year_diff"].values[i] + 0.2, str(df_sorted_asia["year_diff"].values[i]), ha='center')
```
